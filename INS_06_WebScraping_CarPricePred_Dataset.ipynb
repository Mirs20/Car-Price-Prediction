```python
import pandas as pd
import numpy as np
from selenium import webdriver 
import warnings
warnings.filterwarnings("ignore")
import time
import requests

from selenium.common.exceptions import StaleElementReferenceException
from selenium.common.exceptions import ElementNotInteractableException
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import TimeoutException

from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
import re
```

```python
dr=webdriver.Chrome(r"C:\web driver\New Folder\chromedriverupdated2.exe")
dr.get('https://www.cardekho.com/used-cars+in+india')
dr.maximize_window()
time.sleep(5)
```

```python
locations=[]
loc_tag=dr.find_elements('xpath',"//li[@class='gsc_col-xs-4 gsc_col-sm-3 gsc_col-md-2']/a")
for i in loc_tag:
    locations.append(i.get_attribute('href'))
locations
```

```python
dr.close()
```

```python

Dict1={}
columns=["car_urls",'brand','city','variant','price','run','fuel','Trans']
for i in columns:
    Dict1[i]=[]
dr1=webdriver.Chrome(r"C:\web driver\New Folder\chromedriverupdated2.exe")    
for loc in locations[:4]:
    dr1.get(loc)
    dr1.maximize_window()
    time.sleep(3)


    for num in range(10):
        dr1.execute_script("window.scrollBy(0,1000)")
        current_url = dr1.current_url    
        time.sleep(2)
        
        tag1= dr1.find_elements("xpath","//div[@class='gsc_col-xs-7  carsName']/a")
        tag2=dr1.find_elements("xpath","//div[@class='gsc_col-xs-7  carsName']/a")
        tag3=dr1.find_elements('xpath',"//div[@class='car_variant']")
        tag4=dr1.find_elements('xpath',"//span[@class='amnt ']")
        tag5=dr1.find_elements('xpath',"//div[@class='truncate dotlist-2']/div[1]")
        tag6=dr1.find_elements('xpath',"//div[@class='truncate dotlist-2']/div[2]")   
        tag7=dr1.find_elements('xpath',"//div[@class='truncate dotlist-2']/div[3]")

        for i,j,k,l,m,n,o in zip(tag1,tag2,tag3,tag4,tag5,tag6,tag7):
            Dict1['car_urls'].append(i.get_attribute('href'))
            Dict1['brand'].append(j.text)
            Dict1['city'].append(current_url.split('+')[-1])
            Dict1['variant'].append(k.text)
            Dict1['price'].append(l.text)
            Dict1['run'].append(m.text)
            Dict1['fuel'].append(n.text)
            Dict1['Trans'].append(o.text)

dr1.close()   
```

```python
df1=pd.DataFrame(Dict1)
print(df1.shape)
df1.head()
```

```python
df1.to_csv(r"C:\Users\A\Documents\flip robo\created\Dataset/CarPricePrediction_Dataset1.csv")
df1.shape
```
